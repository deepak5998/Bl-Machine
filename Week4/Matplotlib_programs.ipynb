```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

class matplotlib_programs:
    
    def __init__(self):
        print("")
    
    def draw_line_with_suitable_label_x_axis_y_axis(self):
        x = np.array([1, 2])
        y = pd.Series([3, 4])
        plt.plot(x, y)
        plt.show()
    
    def draw_line_using_given_axis_values(self):
        x = np.array(np.arange(5))
        y = np.array(np.random.rand(5))
        plt.plot(x, y)
        plt.title('With x and y labels')
        plt.xlabel('x axis')
        plt.ylabel('y axis')
        plt.show()

    def line_from_text_file(self):
        x, y = np.loadtxt('test.txt', unpack=True, delimiter=' ')
        plt.plot(x, y)
        plt.show()
    
    def plot_two_more_lines_on_same_plot(self):
        x = np.array(np.arange(8))
        y = np.array(np.random.rand(8))
        plt.plot(x, y, 'r-')
        x1 = np.array(np.random.rand(8))
        y1 = np.array(np.random.rand(8))
        plt.plot(x1, y1, 'b')
        plt.plot(x, x1, 'g--') 

matplotlib_obj = matplotlib_programs()
# matplotlib_obj.draw_line_with_suitable_label_x_axis_y_axis()
# matplotlib_obj.draw_line_using_given_axis_values()
# matplotlib_obj.line_from_text_file()
matplotlib_obj.plot_two_more_lines_on_same_plot()   
```
